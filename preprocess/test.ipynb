```python
import pandas as pd
```

```python
stock_data=pd.read_csv('./stock.csv')
stock_values=stock_data[stock_data.symbol=='GOOG'].price.to_list()
stock_values
```

```python
turning_points=[]
turning_pointsA=[]
turning_pointsV=[]
for i in range(1,len(stock_values)-1):
	if stock_values[i]>stock_values[i+1] and stock_values[i]>stock_values[i-1]:
		turning_points.append(stock_values[i])
		turning_pointsA.append(stock_values[i])
	if stock_values[i]<stock_values[i+1] and stock_values[i]<stock_values[i-1]:
		turning_points.append(stock_values[i])
		turning_pointsV.append(stock_values[i])
# turning_points
```
